# Convergence scatterplot

```stata
* Load dataset
use "india520.dta", clear
```

```stata
regress light_growth96_10rcr_cap log_light96_rcr_cap, robust
```

```stata
*------------------------------------------------------------------------------
* (4.1) Unconditional Convergence Plot
*------------------------------------------------------------------------------
scatterfit light_growth96_10rcr_cap log_light96_rcr_cap, ///
    regparameters(coef sig r2) ///
    vce(cluster POLY_ID) ///
    parpos(0.15 -3) ///
    plotscheme(white_tableau) ///
    opts(legend(off) subtitle("") ///
         graphregion(margin(zero)) ///
         scale(1.3) ///
         name(fig1, replace))
```
